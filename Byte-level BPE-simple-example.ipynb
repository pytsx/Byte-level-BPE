### Byte-level BPE
##### Iniciando o tokenizador e o córpus de treinamento (wikipedia)

```python
from tokenizers import ByteLevelBPETokenizer
```

##### Treinando o Tokenizador

```python
tokenizerDPE = ByteLevelBPETokenizer()
files = [f"wikitext-103-raw/wiki.{split}.raw" for split in ['test', 'train', 'valid']]
```

```python
tokenizerDPE.train(files=files, vocab_size=52_000, min_frequency=2, special_tokens=[
    "<s>",
    "<pad>",
    "</s>",
    "<unk>",
    "<mask>",
])
```

```python
tokenizerDPE.save("tokenizer-wiki-byte-level-BPE.json")
```

```python
outoutDPE = tokenizerDPE.encode("No início de julho, em um calor sufocante, às quatro horas da tarde, um jovem saiu do quartinho alugado de uma casa de dois andares no beco de S. e caminhou lentamente, como se estivesse hesitando, em direção à ponte de K. Ele era um jovem de cerca de vinte e cinco anos, magro e baixo, de cabelos castanhos, rosto pálido e que parecia doente. Vestia uma jaqueta de verão e um boné de palha velho, e carregava uma mochila grande nas costas.")
outoutDPE.tokens
```
